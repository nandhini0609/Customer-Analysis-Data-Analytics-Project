```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df
```

```python
#create a column
labels=['Young Adult','Adults','Middle-Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
df[['age','age_group']].head(10)
```

```python
#creating a column for frequently purchasing dates
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Every 3 Months':90,
    'Monthly':30
}
df['purchase_frequent_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequent_days','frequency_of_purchases']]
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df=df.drop('promo_code_used',axis=1)
```

```python
import pyodbc
from sqlalchemy import create_engine

conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=localhost\SQLEXPRESS;"
    "DATABASE=customer_behaviour;"
    "Trusted_Connection=yes;"
)

print("Connected Successfully!")
```

```python
server = r'localhost\SQLEXPRESS'
database = 'customer_behaviour'

engine = create_engine(
    f"mssql+pyodbc://@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"
)

print("Engine created successfully!")
```

```python
df.head()
```

```python
df.to_sql(
    name='customer_data',   # Table name in SQL Server
    con=engine,
    if_exists='replace',    # Creates/replaces the table
    index=False
)

print("Table uploaded successfully!")
```
